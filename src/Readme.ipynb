# Toxic Comment Classification

**Goal:** Build a multi-label text classifier to detect toxic content in online comments, including hate speech, threats, insults, and obscene language.

**Dataset:** [Jigsaw Toxic Comment Classification Challenge](https://www.kaggle.com/c/jigsaw-toxic-comment-classification-challenge) — ~160k Wikipedia comments labeled across 6 toxicity categories.

**Approach:** We start with a TF-IDF + Logistic Regression baseline, get a performance benchmark before investing in heavier approaches like transformers.
